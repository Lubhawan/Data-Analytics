```python
import pandas as pd
import regex as re
# import numpy as np
# from openpyxl import load_workbook

def extract_columns_and_values(file_path, target_column_names):
    """
    Extract specific columns and their values from an Excel file where column names 
    can be located anywhere in the sheet.
    
    Parameters:
    file_path (str): Path to the Excel file
    target_column_names (list): List of column names to extract
    
    Returns:
    dict: Dictionary with column names as keys and their values as lists
    """
    # Load the entire Excel file as a DataFrame without headers
    df = pd.read_excel(file_path, header=None)
    
    # Convert to numpy array for easier searching
    data_array = df.values
    
    result = {}
    
    # Initialize with empty lists for each target column
    for col_name in target_column_names:
        result[col_name] = []
    
    # Search for each target column name in the entire sheet
    for col_name in target_column_names:
        # Find all occurrences of the column name in the sheet
        matches = []
        for i in range(data_array.shape[0]):
            for j in range(data_array.shape[1]):
                cell_value = data_array[i, j]
                # Handle NaN values and compare with column name
                print(cell_value)
                if isinstance(cell_value, str) and re.search(col_name,cell_value,re.IGNORECASE):
                    matches.append((i, j))
        
        # For each occurrence of this column name
        for row_idx, col_idx in matches:
            # Extract values below this column name
            values = []
            for r in range(row_idx + 1, data_array.shape[0]):
                cell_value = data_array[r, col_idx]
                # Stop when we hit an empty cell or NaN
                if pd.isna(cell_value) or cell_value == '':
                    break
                values.append(cell_value)
            
            # Add these values to the result
            result[col_name].extend(values)
    
    return result

def main():
    # Example usage
    file_path = r"C:\Users\Lubhawan Parihar\Downloads\CPR-39014 Metallus Inc - 212126 - 010125 - New Group Pricing Set Up Mixer Form.xlsx"
    target_columns = ["Member\sproduct"]  # Replace with your target column names
    print(target_columns)
    
    try:
        extracted_data = extract_columns_and_values(file_path, target_columns)
        print(extracted_data)
        
        # Print the extracted data
        for col_name, values in extracted_data.items():
            print(f"\n{col_name}:")
            for value in values:
                print(f"  - {value}")
        
        # Optionally, save the extracted data to a new Excel file
        output_df = pd.DataFrame(extracted_data)
        output_df.to_excel("extracted_data.xlsx", index=False)
        print("\nExtracted data saved to 'extracted_data.xlsx'")
        
    except Exception as e:
        print(f"An error occurred: {e}")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import argparse
import os
import re
from difflib import get_close_matches

def extract_columns(input_file, output_file=None, columns_to_extract=None):
    """
    Extract specific columns from an Excel file.
    
    Parameters:
        input_file (str): Path to the input Excel file
        output_file (str, optional): Path to save the extracted data
        columns_to_extract (list, optional): List of column names to extract
                                           If None, extracts all mandatory columns
    
    Returns:
        pd.DataFrame: DataFrame containing the extracted columns
    """
    # Default columns to extract (mandatory columns marked with *)
    default_columns = [
        "Member contract code",
        "Source System",
        "Member Product",
        "Coverage",
        "Line of Business",
        "Contracting State",
        "Tiered Product",
        "Reference mixer",
        "Special Handling",
        "Product Name",
        "Network",
        "AltNet",
        "Npar Facility",
        "NPAR Ambulance",
        "NPAR Facility Emergency Room",
        "NPAR Emergency Claim Inst",
        "N/Par Default",
        "Npar Professional",
        "Prof N/Par",
        "Npar RAPS",
        "Npar AERP",
        "Npar Emergency Claims",
        "Npar Assistant Surgeon",
        "Npar Ambulance",
        "Onsite Clinic",
        "Online Providers",
        "Sydney K-Chat",
        "Lark",
        "ASH",
        "Type of mixer work",
        "Special Instructions"
    ]
    
    # Mandatory columns (marked with * in the description)
    mandatory_columns = [
        "Member contract code",
        "Member Product",
        "Reference mixer",
        "Special Handling",
        "Product Name",
        "Network"
    ]
    
    try:
        # Read Excel file
        print(f"Reading Excel file: {input_file}")
        df = pd.read_excel(input_file)
        
        # Clean column names to handle garbage values
        df.columns = [clean_column_name(col) for col in df.columns]
        
        # Match columns with the desired ones, even when they have extra characters
        column_mapping = {}
        for col in df.columns:
            cleaned_col = clean_column_name(col)
            best_match = find_best_match(cleaned_col, default_columns)
            if best_match:
                column_mapping[col] = best_match
                print(f"Matched column '{col}' to '{best_match}'")
        
        # Rename columns in the dataframe
        if column_mapping:
            df = df.rename(columns=column_mapping)
        
        # Determine which columns to extract
        if columns_to_extract is None:
            columns_to_extract = mandatory_columns
        
        # Check for column existence and create a new dataframe with only the desired columns
        columns_found = []
        for col in columns_to_extract:
            if col in df.columns:
                columns_found.append(col)
            else:
                # Try to find a close match for this column
                possible_matches = [c for c in df.columns if col.lower() in c.lower()]
                if possible_matches:
                    best_match = possible_matches[0]
                    columns_found.append(best_match)
                    print(f"Using '{best_match}' for requested column '{col}'")
                else:
                    print(f"Warning: Column '{col}' not found in the Excel file")
        
        if not columns_found:
            print("Error: None of the specified columns were found in the Excel file")
            return None
        
        # Extract only the columns we want
        extracted_df = df[columns_found]
        
        # Save to a new Excel file if output_file is provided
        if output_file:
            directory = os.path.dirname(output_file)
            if directory and not os.path.exists(directory):
                os.makedirs(directory)
            
            extracted_df.to_excel(output_file, index=False)
            print(f"Extracted data saved to: {output_file}")
        
        return extracted_df
    
    except Exception as e:
        print(f"Error processing the Excel file: {str(e)}")
        return None

def clean_column_name(column_name):
    """
    Clean column name by removing garbage values and extra characters.
    
    Parameters:
        column_name (str): Original column name
        
    Returns:
        str: Cleaned column name
    """
    if not isinstance(column_name, str):
        return str(column_name)
    
    # Remove non-alphanumeric characters except spaces and /
    cleaned = re.sub(r'[^\w\s/]', '', column_name)
    
    # Remove extra spaces
    cleaned = re.sub(r'\s+', ' ', cleaned).strip()
    
    return cleaned

def find_best_match(column_name, target_columns, threshold=0.6):
    """
    Find the best match for a column name from a list of target columns.
    
    Parameters:
        column_name (str): Column name to match
        target_columns (list): List of target column names
        threshold (float): Minimum similarity score (0-1) to consider a match
        
    Returns:
        str or None: Best matching column name or None if no match found
    """
    # Try exact match first (case insensitive)
    for target in target_columns:
        if column_name.lower() == target.lower():
            return target
    
    # Try contains match
    for target in target_columns:
        if column_name.lower() in target.lower() or target.lower() in column_name.lower():
            # If one is fully contained in the other, it's a good match
            if len(column_name) >= 0.7 * len(target) or len(target) >= 0.7 * len(column_name):
                return target
    
    # Try fuzzy matching
    matches = get_close_matches(column_name.lower(), [t.lower() for t in target_columns], n=1, cutoff=threshold)
    if matches:
        matched_index = [t.lower() for t in target_columns].index(matches[0])
        return target_columns[matched_index]
    
    return None

def main():
    # Set up argument parser
    parser = argparse.ArgumentParser(description='Extract specific columns from an Excel file')
    parser.add_argument('input_file', help='Path to the input Excel file')
    parser.add_argument('-o', '--output_file', help='Path to save the extracted data (optional)')
    parser.add_argument('-c', '--columns', nargs='+', help='List of column names to extract (optional)')
    
    # Parse arguments
    args = parser.parse_args()
    
    # Extract columns
    extract_columns(r"C:\Users\Lubhawan Parihar\Downloads\CPR-39014 Metallus Inc - 212126 - 010125 - New Group Pricing Set Up Mixer Form.xlsx", args.output_file, args.columns)

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import argparse
import os
import re
from difflib import get_close_matches
import logging

# Set up logging
logging.basicConfig(level=logging.INFO, 
                    format='%(asctime)s - %(levelname)s - %(message)s')
logger = logging.getLogger('excel_extractor')

def extract_columns(input_file, output_file=None, columns_to_extract=None, verbose=False):
    """
    Extract specific columns from an Excel file.
    
    Parameters:
        input_file (str): Path to the input Excel file
        output_file (str, optional): Path to save the extracted data
        columns_to_extract (list, optional): List of column names to extract
                                           If None, extracts all mandatory columns
        verbose (bool): Enable verbose output for debugging
    
    Returns:
        pd.DataFrame: DataFrame containing the extracted columns
    """
    # Default columns to extract (mandatory columns marked with *)
    default_columns = [
        "Member contract code",
        "Source System",
        "Member Product",
        "Coverage",
        "Line of Business",
        "Contracting State",
        "Tiered Product",
        "Reference mixer",
        "Special Handling",
        "Product Name",
        "Network",
        "AltNet",
        "Npar Facility",
        "NPAR Ambulance",
        "NPAR Facility Emergency Room",
        "NPAR Emergency Claim Inst",
        "N/Par Default",
        "Npar Professional",
        "Prof N/Par",
        "Npar RAPS",
        "Npar AERP",
        "Npar Emergency Claims",
        "Npar Assistant Surgeon",
        "Npar Ambulance",
        "Onsite Clinic",
        "Online Providers",
        "Sydney K-Chat",
        "Lark",
        "ASH",
        "Type of mixer work",
        "Special Instructions"
    ]
    
    # Mandatory columns (marked with * in the description)
    mandatory_columns = [
        "Member contract code",
        "Member Product",
        "Reference mixer",
        "Special Handling",
        "Product Name",
        "Network"
    ]
    
    try:
        logger.info(f"Reading Excel file: {input_file}")
        
        # Try to read the Excel file with verbose debugging
        try:
            # First: Get all sheet names
            xls = pd.ExcelFile(input_file)
            sheet_names = xls.sheet_names
            logger.info(f"Excel file contains these sheets: {sheet_names}")
            
            # Try each sheet to find relevant data
            df = None
            most_matches = 0
            best_sheet = None
            
            for sheet_name in sheet_names:
                logger.info(f"Examining sheet: '{sheet_name}'")
                
                try:
                    # Read the sheet
                    sheet_df = pd.read_excel(input_file, sheet_name=sheet_name)
                    
                    if verbose:
                        logger.info(f"Sheet '{sheet_name}' has {len(sheet_df.columns)} columns and {len(sheet_df)} rows")
                        logger.info(f"Sample columns: {list(sheet_df.columns)[:5]}")
                    
                    # Look for rows with headers - sometimes data starts after a few rows
                    # We'll analyze the first 20 rows to find potential headers
                    for start_row in range(min(20, len(sheet_df))):
                        # Create a new DataFrame starting from this row
                        header_df = sheet_df.iloc[start_row:].copy()
                        
                        # Use this row as the header
                        if len(header_df) > 0:  # Make sure we have data
                            header_row = header_df.iloc[0]
                            
                            # Check if this row contains any column keywords
                            keywords_found = sum(1 for val in header_row if isinstance(val, str) and 
                                               any(keyword.lower() in val.lower() for keyword in 
                                                   ["member", "product", "network", "npar", "mixer"]))
                            
                            if keywords_found >= 3:  # If we found several keywords, use this as header
                                logger.info(f"Found potential header row at index {start_row}")
                                # Set this row as header and skip it in the data
                                header_df.columns = header_df.iloc[0]
                                header_df = header_df.iloc[1:].reset_index(drop=True)
                                
                                # Count potential matches in this configuration
                                potential_matches = 0
                                for col in header_df.columns:
                                    for target in default_columns:
                                        if isinstance(col, str) and (target.lower() in col.lower() or col.lower() in target.lower()):
                                            potential_matches += 1
                                            break
                                
                                if potential_matches > most_matches:
                                    most_matches = potential_matches
                                    best_sheet = sheet_name
                                    df = header_df
                                    logger.info(f"Sheet '{sheet_name}' with header at row {start_row} has {potential_matches} potential column matches")
                    
                    # If we didn't find a good header row, check the original columns
                    if best_sheet is None or best_sheet != sheet_name:
                        potential_matches = 0
                        for col in sheet_df.columns:
                            for target in default_columns:
                                if isinstance(col, str) and (target.lower() in col.lower() or col.lower() in target.lower()):
                                    potential_matches += 1
                                    break
                        
                        if potential_matches > most_matches:
                            most_matches = potential_matches
                            best_sheet = sheet_name
                            df = sheet_df
                            logger.info(f"Sheet '{sheet_name}' with original headers has {potential_matches} potential column matches")
                
                except Exception as e:
                    logger.warning(f"Error processing sheet '{sheet_name}': {str(e)}")
            
            if df is None:
                # If no sheet had obvious matches, use the first sheet
                df = pd.read_excel(input_file, sheet_name=0)
                logger.warning("No obvious matching sheet found. Using the first sheet.")
            else:
                logger.info(f"Selected sheet '{best_sheet}' with {most_matches} potential column matches")
        
        except Exception as sheet_error:
            logger.error(f"Error reading sheets: {str(sheet_error)}")
            # Fallback: try reading without specifying a sheet
            df = pd.read_excel(input_file)
            
        # Debug: show data
        if verbose:
            logger.info("\nFirst 5 rows of data:")
            logger.info(df.head())
            logger.info("\nAll column names found in the file:")
            logger.info(df.columns.tolist())

        # Attempt to handle non-string column names
        df.columns = [str(col) for col in df.columns]
        
        # Clean column names to handle garbage values
        df.columns = [clean_column_name(col) for col in df.columns]
        
        # Check for Excel files with dropdown/choice cells in the header
        # These often contain the format "Column Name - Select..." or similar patterns
        df.columns = [col.split(' - ')[0].split(':')[0].strip() for col in df.columns]
        
        # Match columns with the desired ones, even when they have extra characters
        column_mapping = {}
        for col in df.columns:
            cleaned_col = clean_column_name(col)
            best_match = find_best_match(cleaned_col, default_columns)
            if best_match:
                column_mapping[col] = best_match
                logger.info(f"Matched column '{col}' to '{best_match}'")
        
        # Rename columns in the dataframe
        if column_mapping:
            df = df.rename(columns=column_mapping)
        
        # Determine which columns to extract
        if columns_to_extract is None:
            columns_to_extract = mandatory_columns
        
        # Check for column existence and create a new dataframe with only the desired columns
        columns_found = []
        for col in columns_to_extract:
            if col in df.columns:
                columns_found.append(col)
            else:
                # Try to find a close match for this column
                possible_matches = [c for c in df.columns if col.lower() in c.lower() or c.lower() in col.lower()]
                if possible_matches:
                    best_match = possible_matches[0]
                    columns_found.append(best_match)
                    logger.info(f"Using '{best_match}' for requested column '{col}'")
                else:
                    logger.warning(f"Column '{col}' not found in the Excel file")
        
        if not columns_found:
            logger.error("None of the specified columns were found in the Excel file")
            return None
        
        # Extract only the columns we want
        extracted_df = df[columns_found]
        
        # Save to a new Excel file if output_file is provided
        if output_file:
            directory = os.path.dirname(output_file)
            if directory and not os.path.exists(directory):
                os.makedirs(directory)
            
            extracted_df.to_excel(output_file, index=False)
            logger.info(f"Extracted data saved to: {output_file}")
        
        return extracted_df
    
    except Exception as e:
        logger.error(f"Error processing the Excel file: {str(e)}")
        return None

def clean_column_name(column_name):
    """
    Clean column name by removing garbage values and extra characters.
    
    Parameters:
        column_name (str): Original column name
        
    Returns:
        str: Cleaned column name
    """
    if not isinstance(column_name, str):
        return str(column_name)
    
    # Remove non-alphanumeric characters except spaces and /
    cleaned = re.sub(r'[^\w\s/]', '', column_name)
    
    # Remove extra spaces
    cleaned = re.sub(r'\s+', ' ', cleaned).strip()
    
    return cleaned

def find_best_match(column_name, target_columns, threshold=0.6):
    """
    Find the best match for a column name from a list of target columns.
    
    Parameters:
        column_name (str): Column name to match
        target_columns (list): List of target column names
        threshold (float): Minimum similarity score (0-1) to consider a match
        
    Returns:
        str or None: Best matching column name or None if no match found
    """
    if not column_name or column_name.isspace():
        return None
        
    # Try exact match first (case insensitive)
    for target in target_columns:
        if column_name.lower() == target.lower():
            return target
    
    # Try contains match
    for target in target_columns:
        if column_name.lower() in target.lower() or target.lower() in column_name.lower():
            # If one is fully contained in the other, it's a good match
            if len(column_name) >= 0.7 * len(target) or len(target) >= 0.7 * len(column_name):
                return target
    
    # Try fuzzy matching
    matches = get_close_matches(column_name.lower(), [t.lower() for t in target_columns], n=1, cutoff=threshold)
    if matches:
        matched_index = [t.lower() for t in target_columns].index(matches[0])
        return target_columns[matched_index]
    
    return None

def main():
    # Set up argument parser
    parser = argparse.ArgumentParser(description='Extract specific columns from an Excel file')
    parser.add_argument('input_file', help='Path to the input Excel file')
    parser.add_argument('-o', '--output_file', help='Path to save the extracted data (optional)')
    parser.add_argument('-c', '--columns', nargs='+', help='List of column names to extract (optional)')
    parser.add_argument('-v', '--verbose', action='store_true', help='Enable verbose output for debugging')
    
    # Parse arguments
    args = parser.parse_args()
    
    # Extract columns
    extract_columns(r"C:\Users\Lubhawan Parihar\Downloads\CPR-39014 Metallus Inc - 212126 - 010125 - New Group Pricing Set Up Mixer Form.xlsx", args.output_file, args.columns, args.verbose)

if __name__ == "__main__":
    main()
```
