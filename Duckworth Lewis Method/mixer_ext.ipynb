```python
import pandas as pd
import re

class ExcelDataProcessor:
    def __init__(self, file_path, sheet_name='Mixer Form'):
        """
        Initialize the ExcelDataProcessor with a file path and sheet name.
        
        Args:
            file_path (str): Path to the Excel file
            sheet_name (str): Name of the sheet to process
        """
        self.file_path = file_path
        self.sheet_name = sheet_name
        self.df = None
        self.header_df = None
        self.extracted_df = None
        
    def load_data(self):
        """Load the Excel file into a pandas DataFrame."""
        self.df = pd.read_excel(self.file_path, sheet_name=self.sheet_name)
        return self
    
    def find_header_row(self, max_rows=100):
        """
        Find the header row in the DataFrame by looking for rows containing
        specific keywords related to member data.
        
        Args:
            max_rows (int): Maximum number of rows to search
            
        Returns:
            ExcelDataProcessor: Self for method chaining
        """
        self.header_df = None
        
        for start_row in range(min(max_rows, len(self.df))):
            temp_df = self.df.iloc[start_row:].copy()
            
            if len(temp_df) > 0:
                header_row = temp_df.iloc[0]
                
                keywords_found = sum(1 for val in header_row if 
                                    isinstance(val, str) and 
                                    any(keyword.lower() in val.lower() for keyword in 
                                        ["member", "product", "network", "npar", "mixer", "special"]))
                
                if keywords_found >= 4:
                    temp_df.columns = temp_df.iloc[0]
                    self.header_df = temp_df.iloc[1:].reset_index(drop=True)
                    break
        
        return self
    
    def find_case_data(self):
        """
        Find rows containing case data by searching for 'case' in the first column
        and filtering out legend/additional information rows.
        
        Returns:
            ExcelDataProcessor: Self for method chaining
        """
        if self.header_df is None:
            raise ValueError("Header row not found. Call find_header_row method first.")
            
        # Find case row
        find_case_row = self.header_df[self.header_df.columns[0]].str.contains('case', case=False)
        find_case_row.dropna(inplace=True)
        
        if not find_case_row.any():
            raise ValueError("No case data found in the sheet.")
            
        first_true_idx = find_case_row[find_case_row].index[0]
        header_df_filtered = self.header_df.loc[first_true_idx:].reset_index(drop=True)
        
        # Filter out legend and additional information rows (those with many NaNs)
        mask = header_df_filtered.isna().sum(axis=1) > 15
        
        if mask.any():
            first_match_idx = mask[mask].index[0]
            self.header_df_filtered = header_df_filtered.loc[:first_match_idx-1].reset_index(drop=True)
        else:
            self.header_df_filtered = header_df_filtered.copy()
            
        return self
    
    def extract_columns(self, keywords):
        """
        Extract specific columns based on provided keywords.
        
        Args:
            keywords (list): List of keywords to match column names
            
        Returns:
            ExcelDataProcessor: Self for method chaining
        """
        if self.header_df_filtered is None:
            raise ValueError("Case data not found. Call find_case_data method first.")
            
        matching_columns = self._find_matching_columns(self.header_df_filtered, keywords)
        
        # Extract only these columns
        self.extracted_df = self.header_df_filtered[matching_columns].copy()
        self.extracted_df.columns = keywords
        
        return self
    
    def _find_matching_columns(self, df, keywords):
        """
        Find columns in the DataFrame that match the provided keywords.
        
        Args:
            df (DataFrame): DataFrame to search for columns
            keywords (list): List of keywords to match
            
        Returns:
            list: List of matching column names
        """
        selected_columns = []

        for keyword in keywords:
            for col in df.columns:
                col_lower = str(col).lower()
                keyword_lower = keyword.lower()

                # Handle specific matches
                if keyword_lower == 'member product' and 'member product' in col_lower:
                    selected_columns.append(col)
                    break
                elif keyword_lower == 'varn' and 'varn' in col_lower:
                    selected_columns.append(col)
                    break
                elif keyword_lower == 'contracting_state' and 'contracting state' in col_lower:
                    selected_columns.append(col)
                    break
                elif keyword_lower == 'special instructions' and 'special instructions' in col_lower:
                    selected_columns.append(col)
                    break
        
        return selected_columns
    
    def extract_effective_dates(self):
        """
        Extract effective dates from the special instructions column.
        
        Returns:
            ExcelDataProcessor: Self for method chaining
        """
        if self.extracted_df is None:
            raise ValueError("Columns not extracted. Call extract_columns method first.")
            
        date_patterns = [
            r'(?:production\s*)?effective\s*date\s*:?\s*(\d{1,2}/\d{1,2}/\d{2,4})',  # MM/DD/YYYY format
            r'(?:production\s*)?effective\s*date\s*:?\s*(\d{2,4}-\d{1,2}-\d{1,2})',  # YYYY-MM-DD format
            r'(?:production\s*)?effective\s*date\s*:?\s*(\d{1,2}-\d{1,2}-\d{2,4})',  # MM-DD-YYYY format
            r'(?:production\s*)?effective\s*date\s*:?\s*(\d{2}[a-zA-Z]{3}\d{2,4})'   # DDMMMYYYY format
        ]
        
        effective_dates = []
        
        for i in range(self.extracted_df.shape[0]):
            for pattern in date_patterns:
                date_match = re.search(pattern, str(self.extracted_df.loc[i]['special instructions']), re.IGNORECASE)
                if date_match:
                    effective_dates.append(date_match.group(1))
                    break
            else:
                effective_dates.append(None)  # No match found
        
        self.extracted_df.loc[:, 'effective date'] = effective_dates
        
        return self
    
    def extract_revenue_codes(self):
        """
        Extract revenue codes from the special instructions column.
        
        Returns:
            ExcelDataProcessor: Self for method chaining
        """
        if self.extracted_df is None:
            raise ValueError("Columns not extracted. Call extract_columns method first.")
            
        revenue_code_pattern = r'Revenue\s*CODE\s*=\s*([0-9/]+)'
        revenue_codes = []
        
        for i in range(self.extracted_df.shape[0]):
            rev_code_match = re.search(revenue_code_pattern, str(self.extracted_df.loc[i]['special instructions']), re.IGNORECASE)
            if rev_code_match:
                revenue_codes.append(rev_code_match.group(1))
            else:
                revenue_codes.append(None)
        
        self.extracted_df.loc[:, 'revenue codes'] = revenue_codes
        
        return self
    
    def get_result(self):
        """Return the processed DataFrame."""
        return self.extracted_df
```

```python
# Example usage
processor = ExcelDataProcessor(r"C:\Users\Lubhawan Parihar\Downloads\PRJ2090836 - PRJ2090816 - SSCR Kaiser Aluminum Corporation - OON Ambulance - OON Urgent Cares.xlsx")
result_df = (processor
             .load_data()
             .find_header_row()
             .find_case_data()
             .extract_columns(['member product', 'varn', 'contracting_state', 'special instructions'])
             .extract_effective_dates()
             .get_result())

print(result_df)
```
